# Bird distributions

## Introduction
In this notebook, we will explore and visualize interesting facts about Snow Geese using line charts, scatterplots, and bar charts.

## Dataset
We will be using a dataset birds.csv that contains information about Snow Geese sightings. The dataset Name	ScientificName	Category	Order	Family	Genus	ConservationStatus	MinLength	MaxLength	MinBodyMass	MaxBodyMass	MinWingspan	MaxWingspan

## Data Exploration
## Load the Data
Let's start by loading the Snow Geese data into a pandas DataFrame and taking a quick look at the data.

```python
import pandas as pd

# Load data
data = pd.read_csv('birds.csv')

# Display the first few rows of the data
data.head()
```

## Visualizing Migration Patterns
Line Chart: Number of Sightings over Time
We can start by visualizing the number of Snow Geese sightings over time to observe their migration patterns.

```python
import matplotlib.pyplot as plt
# Line Chart: Conservation Status
conservation_counts = data['ConservationStatus'].value_counts()

plt.figure(figsize=(10, 6))
plt.plot(conservation_counts.index, conservation_counts.values)
plt.title("Bird Conservation Status")
plt.xlabel("Conservation Status")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

## Scatterplot: Body Mass vs. Wingspan

```python
plt.figure(figsize=(10, 6))
plt.scatter(data['MaxBodyMass'], data['MaxWingspan'])
plt.title("Bird Body Mass vs. Wingspan")
plt.xlabel("Max Body Mass")
plt.ylabel("Max Wingspan")
plt.grid(True)
plt.show()
```

## Bar Chart: Bird Categories

```python
category_counts = data['Category'].value_counts()

plt.figure(figsize=(10, 6))
plt.bar(category_counts.index, category_counts.values)
plt.title("Bird Categories")
plt.xlabel("Category")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```
